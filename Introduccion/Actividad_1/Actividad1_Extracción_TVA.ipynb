## 1Ô∏è‚É£ JSON: Informaci√≥n de empleados

Sup√≥n que tienes el archivo <code>empleados.json</code> con la siguiente estructura:

```json
[
  {"nombre": "Ana", "edad": 28, "departamento": "Ventas", "salario": 2500},
  {"nombre": "Luis", "edad": 35, "departamento": "IT", "salario": 3200},
  {"nombre": "Mar√≠a", "edad": 24, "departamento": "Ventas", "salario": 2300},
  {"nombre": "Carlos", "edad": 40, "departamento": "IT", "salario": 4000}
]
```

<b>Ejercicios:</b>

1. Crea un DataFrame a partir de la extracci√≥n de informaci√≥n de este archivo JSON.
2. ¬øCu√°l es el <b>salario promedio por departamento</b>?
3. ¬øQui√©n es el <b>empleado m√°s joven</b> y a qu√© departamento pertenece?
4. ¬øCu√°ntos empleados tienen un <b>salario mayor a 2500</b>?
5. ¬øCu√°l es la <b>diferencia de edad</b> entre el empleado m√°s joven y el de mayor edad?
6. ¬øCu√°l es el <b>salario total pagado</b> en la empresa?

<i>Tip: Usa pandas.read_json o json.load para cargar los datos.</i>

````python
[
  {"nombre": "Ana", "edad": 28, "departamento": "Ventas", "salario": 2500},
  {"nombre": "Luis", "edad": 35, "departamento": "IT", "salario": 3200},
  {"nombre": "Mar√≠a", "edad": 24, "departamento": "Ventas", "salario": 2300},
  {"nombre": "Carlos", "edad": 40, "departamento": "IT", "salario": 4000}
]
````

````python
import pandas as pd
import json

# cARGO ARCHIVO
data = [
    {"nombre": "Ana", "edad": 28, "departamento": "Ventas", "salario": 2500},
    {"nombre": "Luis", "edad": 35, "departamento": "IT", "salario": 3200},
    {"nombre": "Mar√≠a", "edad": 24, "departamento": "Ventas", "salario": 2300},
    {"nombre": "Carlos", "edad": 40, "departamento": "IT", "salario": 4000}
]

df = pd.DataFrame(data)

# cREAR EL DATA FRAME
print("DataFrame de empleados:")
print(df)


````

````python
# CALCULAR SALARIO
prom_sal = df.groupby("departamento")["salario"].mean()
print("\nSalario promedio por cada departamento:")
print(prom_sal)

````

````python
# IDETIFICAR EMPLEADOS MAS JOVENES Y DEPTO
emp_joven = df.loc[df["edad"].idxmin()]
print("\nEmpleado m√°s joven:")
print(f"Nombre: {emp_joven['nombre']}, Departamento: {emp_joven['departamento']}")

````

````python
#EMPLEADOS C SALARIO MAYOR A 2500
mayor_2500 = df[df["salario"] > 2500].shape[0]
print("\nEmpleados con salario mayor a 2500:")
print(mayor_2500)

````

````python
# DIFERENCIA ENTRE MAYOR Y MENOR
dif_edad = df["edad"].max() - df["edad"].min()
print("\nDiferencia de edad entre el mayor y el menor:")
print(dif_edad)
````

````python

# 6. SALARIO PAGADO FINAL
salario_t = df["salario"].sum()
print("\n Salario total pagado por la empresa:")
print(salario_t)
````

Sup√≥n que tienes el archivo <code>libros.xml</code> con la siguiente estructura:

```xml
  </libro>

```

<b>Ejercicios:</b>

1. Crea un DataFrame a partir de la extracci√≥n de informaci√≥n de este archivo XML.
2. ¬øCu√°l es el <b>libro m√°s caro</b> y su autor?
3. ¬øCu√°l es el <b>precio promedio</b> de los libros publicados despu√©s de 2020?
4. ¬øCu√°ntos libros fueron <b>publicados antes de 2021</b>?
5. ¬øCu√°l es la <b>diferencia de precio</b> entre el libro m√°s barato y el m√°s caro?
6. ¬øCu√°ntos <b>autores diferentes</b> hay en la lista?

<i>Tip: Usa xml.etree.ElementTree para leer el archivo y pandas para tabular los datos.</i>

````python
#Alma Trejo
import xml.etree.ElementTree as ET
import pandas as pd

tree = ET.parse('libros.xml')   
root = tree.getroot()

libros = []
for libro in root.findall('libro'):
    titulo = libro.find('titulo').text if libro.find('titulo') is not None else None
    autor = libro.find('autor').text if libro.find('autor') is not None else None
    precio = libro.find('precio').text if libro.find('precio') is not None else None
    anio = libro.find('anio').text if libro.find('anio') is not None else None

    
    try:
        precio = float(precio) if precio is not None else None
    except ValueError:
        precio = None
    try:
        anio = int(anio) if anio is not None else None
    except ValueError:
        anio = None

    libros.append({
        "titulo": titulo,
        "autor": autor,
        "precio": precio,
        "anio": anio
    })

df_libros = pd.DataFrame(libros)
print(df_libros)
````

````python
# 1. LIBRO MAS CARO Y AUTOR
libro_caro = df_libros.loc[df_libros['precio'].idxmax()]
print("\nLibro m√°s caro:")
print(f"T√≠tulo: {libro_caro['titulo']}, Autor: {libro_caro['autor']}")
````

````python
# 2. Precio prom de libros publicados despues de 2020
prom_2020 = df_libros[df_libros['anio'] > 2020]['precio'].mean()
print("\nPrecio promedio de libros publicados despu√©s de 2020:")
print(prom_2020)
````

````python

# 3. Libros pub antes de 2021
libros_antes_2021 = df_libros[df_libros['anio'] < 2021].shape[0]
print("\nüìå Libros publicados antes de 2021:")
print(libros_antes_2021)
````

````python

# 4. Diferencia entre precio mas caro y mas baratp
dif_precios = df_libros['precio'].max() - df_libros['precio'].min()
print("\nüìå Diferencia de precio entre el m√°s barato y el m√°s caro:")
print(dif_precios)
````

````python

# 5. Autores diferentes
autores_unicos = df_libros['autor'].nunique()
print("\nüìå Cantidad de autores diferentes:")
print(autores_unicos)
````

## 3Ô∏è‚É£ CSV: Ventas detalladas

Sup√≥n que tienes el archivo <code>ventas_detalladas.csv</code> con la siguiente estructura:

```csv
producto,fecha,cantidad,precio_unitario
A,2023-01-01,10,2.5
B,2023-01-02,5,3.0
A,2023-01-03,7,2.5
C,2023-01-01,3,4.0
B,2023-01-03,2,3.0
```

<b>Ejercicios:</b>

1. Crea un DataFrame con los datos extra√≠dos del archivo.
2. ¬øCu√°l es el <b>total vendido</b> por producto y por fecha?
3. ¬øEn qu√© fecha se vendi√≥ la <b>mayor cantidad de productos</b>?
4. ¬øCu√°l es el <b>producto con mayor ingreso total</b>?
5. ¬øCu√°l es la <b>media de cantidad vendida</b> por producto?
6. ¬øQu√© producto tuvo <b>ventas todos los d√≠as registrados</b>?

<i>Tip: Usa pandas.read_csv y agrupa los datos seg√∫n lo que se pide.</i>

````python
import pandas as pd

# Alma Trejo
df = pd.read_csv("ventas_detalladas.csv")

print("DataFrame cargado:")
print(df)

````

````python

# 1. Total vendido producto/fecha (cantidad * precio_unitario)
df["total"] = df["cantidad"] * df["precio_unitario"]

totales_prod_f = df.groupby(["producto", "fecha"])[["cantidad", "total"]].sum()
print("\nTotal vendido por producto y por fecha:")
print(totales_prod_f)
````

````python

# 2. Fecha con mayor cantidad total vendida
cantidad_por_f = df.groupby("fecha")["cantidad"].sum()
fecha_max = cantidad_por_f.idxmax()

print("\n Fecha con mayor cantidad de productos vendidos:")
print(fecha_max)

````

````python
# 3. Producto con mayor ingreso total
ingreso_por_producto = df.groupby("producto")["total"].sum()
producto_max_ingreso = ingreso_por_producto.idxmax()

print("\n Producto con mayor ingreso total:")
print(producto_max_ingreso)
````

````python
# 4. Media de cantidad vendida por producto
media_cantidad = df.groupby("producto")["cantidad"].mean()
print("\n Media de cantidad vendida por producto:")
print(media_cantidad)

````

````python
# 5. Producto que tuvo ventas todos los d√≠as registrados
# Obtener todas las fechas √∫nicas
fechas_unicas = df["fecha"].nunique()

# Contar en cu√°ntas fechas aparece cada producto
productos_frecuencia = df.groupby("producto")["fecha"].nunique()

productos_todos_dias = productos_frecuencia[productos_frecuencia == fechas_unicas].index.tolist()

print("\n Producto(s) con ventas todos los d√≠as registrados:")
print(productos_todos_dias)
````
